```python
import cv2
import matplotlib.pyplot as plt

# Read an image in BGR format
bgr_image = cv2.imread('img1.jpg')

# Convert BGR to RGB without changing pixel values
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# Display both images side by side
plt.subplot(121)
plt.imshow(bgr_image)
plt.title('BGR Image')

plt.subplot(122)
plt.imshow(rgb_image)
plt.title('RGB Image (Display Only)')

plt.show()
```
