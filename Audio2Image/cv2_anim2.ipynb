```python
import cv2
import wave
import time
import pygame, pygame.mixer
import numpy as np, random
from PIL import Image, ImageEnhance


# load video and image files
video = cv2.VideoCapture("Vid2.mp4")
img = cv2.imread('img1.jpg')
cv2_image=cv2.imread("img1.jpg") # open image using cv2

# load audio file, audio data
aud = wave.open("PinkPanther30.wav")

num_frames = aud.getnframes()
audio_fps = aud.getframerate()
frames = aud.readframes(num_frames) # reads all the frames of the audio
duration = num_frames / audio_fps

print(f"Audio total frames: {aud.getnframes()}")
print(f"Audio duration: {duration}")
print(f"Audio fps: {audio_fps}")

# compressed audio file
aud_array = np.frombuffer(frames, dtype=float, count=-1, offset=0)
aud_comp = aud_array[0:num_frames:10000]

# window name and size, video data
cv2.namedWindow("video", cv2.WINDOW_NORMAL)
length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH)) 
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)
print(f"Video total frames: {length} ")
print(f"Video duration: {length / fps} ")
print(f"Video fps: {fps} ")

# play audio
def wav_play(audio_file):
    pygame.mixer.init()
    pygame.mixer.music.load(audio_file)
    pygame.mixer.music.play()

    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

    pygame.mixer.quit()

audio_file = "PinkPanther30.wav"
# wav_play(audio_file)


while True:
    # Read video capture
    ret, frame = video.read()

    # Convert from cv2 to pillow
    color_converted = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) # convert from BGR to RGB
    pil_image = Image.fromarray(color_converted)

    # Enhance Image
    contrast_enhancer = ImageEnhance.Contrast(pil_image)
    # colour_enhancer = ImageEnhance.Color(pil_image)
    # enhancer = ImageEnhance.Brightness(pil_image)
    # enhancer = ImageEnhance.Sharpness(pil_image)

    for value in range(aud.getnframes()):
        contrast_factor = (abs(frames[value])) # RANDOM: round(random.uniform(0.5 ,5), 2)
    pil_image2 = contrast_enhancer.enhance(contrast_factor)

    # convert from pillow to cv2
    opencv_image = np.array(pil_image2)
    final_image = cv2.cvtColor(opencv_image, cv2.COLOR_RGB2BGRA) # convert from RGB to BGR 

    # CONTRAST ENHANCING WITH CV2:
    # alpha = round(random.uniform(1,3), 1) # Contrast control (1.0-3.0)
    # beta = 0 # random.randint(0,100) # Brightness control (0-100)
    # frame = cv2.convertScaleAbs(frame, alpha=alpha, beta=beta)

    # Display each frame
    cv2.imshow("video", final_image)
    # cv2.waitKey(00) == ord('k')
    if cv2.waitKey(5) == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
aud.close()
```
